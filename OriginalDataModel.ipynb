```python
import tensorflow as tf
from tensorflow import keras

import matplotlib.pyplot as plt
```

```python
train_dir = r"/Users/sparshkansotia/deep_learning/Braindead/Train"

dataset = keras.utils.image_dataset_from_directory(
    directory = train_dir,
    image_size = [180, 180],
    color_mode = "grayscale",
    labels="inferred",
    label_mode = "int"
)

train_size = int(0.7*dataset.cardinality().numpy())
train_dataset = dataset.take(train_size)
val_dataset = dataset.skip(train_size)
```

```python
original_model = keras.models.Sequential([
    keras.layers.Rescaling(1./255, input_shape=[180, 180, 1]),
    keras.layers.Conv2D(32, kernel_size=3, strides=1, padding="same", activation="relu"),
    keras.layers.MaxPooling2D(pool_size=3),
    keras.layers.Conv2D(64, kernel_size=3, strides=1, padding="same", activation="relu"),
    keras.layers.MaxPooling2D(pool_size=2),
    keras.layers.Conv2D(128, kernel_size=3, strides=1, padding="same", activation="relu"),
    keras.layers.MaxPooling2D(pool_size=2),
    keras.layers.Conv2D(256, kernel_size=3, strides=1, padding="same", activation="relu"),
    keras.layers.MaxPooling2D(pool_size=2),
    keras.layers.Flatten(),
    keras.layers.Dense(80, activation="relu"),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(20, activation="relu"),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(4, activation="softmax")
])

cb1 = keras.callbacks.ModelCheckpoint("Original.h5", save_best_only=True)

original_model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
```

```python
history = original_model.fit(train_dataset, epochs=500, validation_data=val_dataset, callbacks=[cb1])
```

```python
generator = keras.models.Sequential([
    keras.layers.Dense(5*5*128, input_shape=[100]),
    keras.layers.Reshape([5, 5, 128]),
    keras.layers.BatchNormalization(),
    keras.layers.Conv2DTranspose(64, kernel_size=3, strides=3, padding="same",activation="selu"),
    keras.layers.BatchNormalization(),
    keras.layers.Conv2DTranspose(32, kernel_size=3, strides=3, padding="same", activation="selu"),
    keras.layers.BatchNormalization(),
    keras.layers.Conv2DTranspose(16, kernel_size=3, strides=2, padding="same", activation="selu"),
    keras.layers.BatchNormalization(),
    keras.layers.Conv2DTranspose(1, kernel_size=5, strides=2, padding="same", activation="tanh")    
])
discriminator = keras.models.Sequential([
    keras.layers.Rescaling(1./255, input_shape=[180, 180, 1]),
    keras.layers.Conv2D(16, kernel_size=5, strides=1, padding="same", activation="selu"),
    keras.layers.MaxPooling2D(pool_size=2),
    keras.layers.Conv2D(32, kernel_size=3, strides=1, padding="same", activation="selu"),
    keras.layers.MaxPooling2D(pool_size=2),
    keras.layers.Conv2D(64, kernel_size=3, strides=1, padding="same", activation="selu"),
    keras.layers.MaxPooling2D(pool_size=2),
    keras.layers.Conv2D(128, kernel_size=3, strides=1, padding="same", activation="selu"),
    keras.layers.MaxPooling2D(pool_size=2),
    keras.layers.Flatten(),
    keras.layers.Dense(80, activation="tanh"),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(24, activation="tanh"),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(1, activation="sigmoid")
])
gan_model = keras.models.Sequential([generator, discriminator])

discriminator.compile(optimizer="adam", loss="binary_crossentropy")
discriminator.trainable=False
gan_model.compile(optimizer="adam", loss="binary_crossentropy")
```

```python
def train_gan(gan_model, dataset, batch_size, codings_size, n_epochs=50):
    generator, discriminator = gan_model.layers
    for epoch in range(n_epochs):
        for X_batch, _ in dataset:
            noise = tf.random.normal(shape=[batch_size, codings_size])
            generated_images = generator(noise)
            X_fake_and_real = tf.concat([generated_images, X_batch], axis=0)
            y1 = tf.constant([[0.]]*batch_size+[[1.]]*batch_size)
            discriminator.trainable=True
            discriminator.train_on_batch(X_fake_and_real, y1)
            noise = tf.random.normal(shape=[batch_size, codings_size])
            y2 = tf.constant([[1.]]*batch_size)
            discriminator.trainable=False
            gan_model.train_on_batch(noise, y2)
```

```python
dir2 = r"F:\PGDBA\Competitions\Braindead\train2\angry2"

real_img = keras.utils.image_dataset_from_directory(
    directory = dir2,
    image_size = [180, 180],
    color_mode = "grayscale",
    labels="inferred",
    label_mode = "int",
    batch_size=8
)

def preprocess(value1, value2):
    return value1*1./255, value2

real_img = real_img.map(preprocess)
real_img = real_img.prefetch(1)
```

```python
train_gan(gan_model=gan_model, dataset=real_img, batch_size=8, codings_size=100, n_epochs=50)
```
